## The Shapefile Format

The format of the flood zone data is provided as *shapefiles*. Shapefiles are a way of encoding geospatial information, developed by the Environmental Systems Research Institute, Inc. (ESRI). In fact, a shapefile consists of three files: a main file (`.shp`), an index file (`.shx`), and a dBASE table (`.dbf`). You can check to see these three files with the same name and respective file extensions in the folder where you unzipped the downloaded `.zip` file into.


In our flood zones example, the shapefile encodes the flood zones as polygons, and it provides the latitude and longitude of the polygon corners. Recall that the latitude of a point on earth indicates how far north or south of the equator it is, and the longitude indicates how much east or west of the prime meridian it is. For reference, Boston is approximately 42.36$^\circ$N, 71.06$^\circ$W. We'll write it as a (Longitude, Latitude) pair, $(-71.06, 42.36)$, where a negative longitude means west of the prime meridian. 

Below is a plot of one such polygon from the "9INCH SEA LEVEL RISE 10PCT ANNUAL FLOOD" shapefile. (For now, you can ignore the code used to generate the plot, by clicking on the toggle button.)


```python
from IPython.display import HTML
HTML(filename="misc/code_toggle.html")
```

```python
import shapefile
import matplotlib.pyplot as plt

sf = shapefile.Reader("data/9inch_Sea_Level_Rise_10pct_Annual_Flood/9inch_Sea_Level_Rise_10pct_Annual_Flood.shp")
x, y = zip(*sf.shape(0).points[sf.shape(0).parts[104] : sf.shape(0).parts[105]])

plt.figure(figsize=(12,15))
plt.axes().set_aspect("equal")
plt.plot(x, y)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

A reader familiar with Boston geography might begin to recognize outlines of some distinctive features of the Boston Harbor: the Charles, Mystic and Neponset Rivers, as well as the Boston Logan Airport. (There are also some unusual outlines that will later become apparent what they are.) As you can see, this is quite a large and complicated polygon. In fact, it is a 155697-sided polygon! 

The shapefile contains more polygons than what's plotted above. Later in this lesson, we will drill down into what the full data actually looks like.

In general, shapefiles can contain a lot of other types of geographic information, such as

- Multiple shapes of various types (points, lines, polygon, etc.). These data are stored in the main `.shp` file.
- Feature attributes of each shape. For example, if a shape represents a county, the feature attributes might contain the county name, population, etc. These data are stored in the dBASE `.dbf` file.

### Understanding the shapefiles

Now, we're ready to load and analyze the shapefiles. If you've hidden the Python code earlier, click the button below now to show the code so you can see what's going on.

```python
HTML(filename="misc/code_show.html")
```

```python
import shapefile
import matplotlib.pyplot as plt

flood_zone = shapefile.Reader("data/9inch_Sea_Level_Rise_10pct_Annual_Flood/9inch_Sea_Level_Rise_10pct_Annual_Flood.shp")
```

The function `Reader` in the `shapefile` module reads the shapefile data into the object `flood_zone`. 

Some useful attributes of `flood_zone` are:

```python
flood_zone.shapeName # Where your data is loaded from
```

```python
flood_zone.shapeType # Shape type 5 is the ESRI shapefile code for polygon
```

```python
flood_zone.numRecords # Number of records in the shapefile
```

There is only 1 *shape* recorded in our shapefile, which in this case is the flood zone itself. In general, a shapefile can contain multiple shapes, such as a shapefile of the US with each State being one shape. What comprises a shape can be fairly arbitrary and dependent on the nature of the geospatial data.

Each shape can comprise of multiple parts. A simple mental picture to understand why there can be many parts is to imagine many lakes in a State---one part is a polygon outlining the State border and the other parts are polygons outlining the interior lakes. Note that the parts are also allowed to be of different shape type.

Returning to our flood zone example, each part in the shape is a polygon. We'll define helper functions that will return the number of parts in a shape and retrieve a part. If you're interested to know how the helper functions work, head to the appendix for details.

```python
# A helper function to return the number of parts in the shape
def get_num_parts(shape):
    return len(shape.parts) - 1

# A helper function to retrieve a polygon part from the shape.
def get_part(shape, part):
    polygon = shape.points[ shape.parts[part] : shape.parts[part+1] ]
    return polygon
```

```python
shape = flood_zone.shape(0) # Get the first shape. Remember, indices start from 0 in Python

nparts = get_num_parts(shape)
"There are " + str(nparts) + " parts in the shape"
```

```python
polygon0 = get_part(shape, 0) # The first part. Again, remember, indices start from 0
polygon0
```

As we can see, `polygon0` comprises of 5 points. Each row of `polygon0` is the (Longitude, Latitude) pair. Note that the first and last point are identical, meaning that it is a 4-sided closed polygon.

Now let us plot the polygon.

```python
# The function zip converts the list of (Longitude, Latitude) pairs into an array of Longitudes and an array of Latitudes
lons0, lats0 = zip(*polygon0)

# Make the plot. 
plt.plot(lons0, lats0)

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

Not a very exciting polygon, is it?

It so happens that the majority of the 856 parts of the flood zone shape are very tiny polygons. Some analysis tells us that the largest polygon is the 105th part. Plotting the 105th part gives the following polygon, which is the same one as shown in the previous section.

```python
polygon104 = get_part(shape, 104) # The 105th part

## Make the plot

plt.axes().set_aspect("equal") # Sets the x- and y-axes to have equal unit length (Q: Why is this necessary?)

lons104, lats104 = zip(*polygon104)
plt.plot(lons104, lats104)

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

Now let's plot the entire shape with all 856 parts. We'll do this using a for loop to loop through each part. For comparison to above, I've plotted the largest polygon in blue.

```python
plt.figure(figsize=(12,15))
plt.axes().set_aspect("equal")

for part in range(nparts):
    polygon = get_part(shape, part)
    lons, lats = zip(*polygon)
    plt.plot(lons, lats, 'k')  # 'k' makes it plot in black

plt.plot(lons104, lats104, 'b') # Overplot the largest polygon in blue ('b')

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

As you can see, the many parts of the shapefile outline islands in Boston Harbor as well as low-lying coastal and riverine regions.

### Overlaying the Flood Zone on Google Maps

The mapping of the flood zone can be made more user-friendly if it is overlaid on an easily recognizable map. Let's do this using Google Maps---it is fast and easy with the Google Maps API and the Python module `gmplot`, and also interactive!

Overlaying the flood zone on Google maps is easy: first, we retrieve the map of Boston. Then, we plot the polygons directly onto the map. Finally, we draw the map into an HTML file for interactive viewing. The following code implements these steps.

```python
import gmplot

# Get a map of Boston
Boston_map = gmplot.GoogleMapPlotter.from_geocode("Boston") # Google Maps API knows where Boston is!

# A technical point of using a Google Maps API authentication key (see below)
fcon = open("./~APIkey_localhost_8888", "r")
Boston_map.apikey = fcon.readline()
fcon.close()

# Plot each polygon part on the map
for part in range(nparts):
    lons, lats = zip(*get_part(shape, part))
    Boston_map.plot(lats, lons, 'b')        # Plot the polygon directly onto the map. Note (lat, lon) not (lon, lat)!

# Draw the map into an HTML file
Boston_map.draw("Boston_Map_With_Flood_Zone.html")
```

*A technical point about authenticating the Google Maps API:* Google requires an authentication key to allow the Javascript applications in the HTML file to access Maps API. I've [created an API key](https://developers.google.com/maps/documentation/javascript/get-api-key#standard-auth) that works only for `localhost:8888` servers, and saved it to a local file named `~APIkey_localhost_8888`. These three lines of code above opens the file, reads the key into the `Boston_map` object, and then closes the file. Remove these three lines of code if you prefer to open the HTML file directly from your local computer.


The map is ready for interactive viewing:

```python
from IPython.display import IFrame      # IFrame embeds the HTML into the iPython document

IFrame("Boston_Map_With_Flood_Zone.html", width = 1000, height = 600)
```

## Comparing Flood Zones

Now that we have a grasp of how to visualize the flood zone for "9INCH SEA LEVEL RISE 10PCT ANNUAL FLOOD", let's move on to comparing flood zones for different sea level rise heights, 9, 21 and 36 inch.

Following the code above, upon loading the shapefiles, we'll plot the 9 inch and 21 inch flood zones together on the same map.

```python
# Get the flood zone shapes
flood_zone_9inch_10pc = shapefile.Reader("data/9inch_Sea_Level_Rise_10pct_Annual_Flood/9inch_Sea_Level_Rise_10pct_Annual_Flood.shp")
shape_9inch_10pc = flood_zone_9inch_10pc.shape(0)

flood_zone_21inch_10pc = shapefile.Reader("data/21inch_Sea_Level_Rise_10pct_Annual_Flood/21inch_Sea_Level_Rise__10pct_Annual_Flood.shp")
shape_21inch_10pc = flood_zone_21inch_10pc.shape(0)
```

```python
# A map of Boston
Boston_map_9v21 = gmplot.GoogleMapPlotter.from_geocode("Boston")

# Authentication key
fcon = open("./~APIkey_localhost_8888", "r")
Boston_map_9v21.apikey = fcon.readline()
fcon.close()

# Plot flood zone for 9inch sea level rise
for part in range(get_num_parts(shape_9inch_10pc)):
    lons, lats = zip(*get_part(shape_9inch_10pc, part))
    Boston_map_9v21.plot(lats, lons, 'b')                # 'b' for blue

# Plot flood zone for 21inch sea level rise
for part in range(get_num_parts(shape_21inch_10pc)):
    lons, lats = zip(*get_part(shape_21inch_10pc, part))
    Boston_map_9v21.plot(lats, lons, 'r')                # 'r' for red
    
Boston_map_9v21.draw("Boston_Map_With_Flood_Zone_9v21.html")
```

## Appendix

### The `get_num_parts` and `get_part` functions

Earlier, we had defined the helper functions `get_num_parts` and `get_part`. Here's an explanation of the shapefile data structure and how those functions work.

Recall that `shape = flood_zone.shape(0)` is the (sole) record in the shapefile, and has all the information about each part of the shape. The two important variables in shape are

- `shape.parts`
- `shape.points`

The latter, `shape.points`, is simply a list of (Longitude, Latitude) pairs. The first 10 points are shown below.

```python
shape.points[:10]
```

Then, `shape.parts` encodes a way to tell us which points belong to which part. The first 4 entries are shown below.

```python
shape.parts[:4]
```

- Part 0 contains the points from 0 (inclusive) up to 5 (exclusive), i.e., `shape.points[0:5]`. 0 is `shape.parts[0]` and 5 is `shape.parts[1]`.
- Part 1 contains the points from 5 (inclusive) up to 10 (exclusive), i.e., `shape.points[5:10]`. 5 is `shape.parts[1]` and 10 is `shape.parts[2]`.
- Part 2 contains the points from 10 (inclusive) up to 15 (exclusive), i.e., `shape.points[10:15]`. 10 is `shape.parts[2]` and 15 is `shape.parts[3]`.

In general, 

> Part `n` contains the points from `shape.parts[n]` (inclusive) up to `shape.parts[n+1]` (exclusive), i.e.,  
>   `shape.points[shape.parts[n] : shape.parts[n+1]]`.

Finally, the total number of parts must be one less than the length of `shape.parts` (otherwise `shape.parts[nparts+1]` will not make sense).
